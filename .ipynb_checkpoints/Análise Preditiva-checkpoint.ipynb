```python
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Salvar modelo de Machine Learning
import joblib

# Matrix de Confusão
from yellowbrick.classifier import ConfusionMatrix 
import pickle

# Acurácia
from sklearn.metrics import accuracy_score
```

***
# <font color=teal>Pipeline
***

```python
# abrir a base criada pelo train_test_split
with open('emprestimo.pkl', mode='rb') as f:
    x_train, y_train, x_test, y_test = pickle.load(f)
```

```python
# abrir a base criada pelo Smote
with open('emprestimo_smote.pkl', mode='rb') as f:
    X_resampled, y_resampled, x_test, y_test = pickle.load(f)
```

```python
# Defina o pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Tratamento de valores nulos (substituição pela média)
    ('scaler', StandardScaler()),  # Pré-processamento (escala dos dados)
    ('pca', PCA(n_components=3)),  # Redução de dimensionalidade com PCA (número de componentes desejados)
    ('xgboost', xgb.XGBClassifier(
        objective='binary:logistic',  # Problema de classificação binária
        max_depth=3,
        learning_rate=0.1,
        n_estimators=100,
        eval_metric='logloss'  # Métrica de avaliação
    ))
])
```

###  Treine o modelo e faça previsões train_test_split

```python
pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)
```

```python
# Avalie o desempenho do modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia: {accuracy:.2f}')
```

```python
# Calcule o RMSE (O RMSE mede o erro médio entre as previsões e os valores reais)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f'RMSE: {rmse:.2f}')

# Calcule o R² (Varia entre 0 e 1. Valores mais próximos de 1 indicam um melhor ajuste do modelo aos dados)
r2 = r2_score(y_test, y_pred)
print(f'R²: {r2:.2f}')
```

### Matrix de Confusão

```python
cm = ConfusionMatrix(pipeline)
cm.fit(x_train, y_train)
cm.score(x_test, y_test)
```

### Treine o modelo e faça previsões Smote

```python
pipeline.fit(X_resampled, y_resampled)
y_pred_2 = pipeline.predict(x_test)
```

```python
# Avalie o desempenho do modelo
accuracy_2 = accuracy_score(y_test, y_pred_2)
print(f'Acurácia: {accuracy:.2f}')
```

```python
# Calcule o RMSE (O RMSE mede o erro médio entre as previsões e os valores reais)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_2))
print(f'RMSE: {rmse:.2f}')

# Calcule o R² (Varia entre 0 e 1. Valores mais próximos de 1 indicam um melhor ajuste do modelo aos dados)
r2 = r2_score(y_test, y_pred_2)
print(f'R²: {r2:.2f}')
```

### Matrix de Confusão

```python
cm = ConfusionMatrix(pipeline)
cm.fit(X_resampled, y_resampled)
cm.score(x_test, y_test)
```

### Conclusão

- Acurácia dos 2 modelo (criou exemplos sintéticos da classe minoritária e ou com classe minoritária), obtiveram o mesmo resultado

- No caso do R² e RMSE com Smote: modelo com a mesma quantidade de registros para ambas as classes, se saiu melhor.
